```python
def convert_to_gray(frames):
    grays = []
    for frame in frames:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        grays.append(gray[None])
    return np.array(grays)

gray_df = frames.map_blocks(
    convert_to_gray,
    chunks=(1, 1, 1080, 1920),
    dtype=np.uint8
)

gray_df.compute()
```

```python
model = KMeans(n_clusters=20, random_state=0)

model.fit(frames)
```
