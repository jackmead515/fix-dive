```python
import cv2
import matplotlib.pyplot as plt
```

```python
frame = cv2.imread('../with_color_features.png')

# rotate 45 degrees

M = cv2.getRotationMatrix2D((frame.shape[1]/2, frame.shape[0]/2), 45, 1)

frame = cv2.warpAffine(
    frame,
    M,
    (frame.shape[1], frame.shape[0]),
    borderMode=cv2.BORDER_REFLECT,
)

plt.imshow(frame)
plt.show()
```
