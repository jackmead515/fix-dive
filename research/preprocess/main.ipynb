```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

url = 'http://minio-api.kubby.ninja/fix-dive-storage/projects/1234567890/features/preprocess/preprocess.gzip.parquet'

df = pd.read_parquet(url)

df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

feature_cols = [col for col in df.columns if col != 'frame_index' and col != 'features']

features = df[feature_cols]

scaler.fit(features)

features = scaler.transform(features)

sdf = pd.DataFrame(features, columns=feature_cols)
sdf['frame_index'] = df['frame_index']
```

```python
# plot blur and total objects
plt.figure(figsize=(20, 10)) 
#plt.scatter(df['frame_index'], df['blur'], color='red')
#plt.scatter(df['frame_index'], df['total_objects'], color='blue')

plt.plot(sdf['frame_index'], sdf['median_motion'].rolling(5).mean(), color='green')
plt.plot(sdf['frame_index'], sdf['blur'].rolling(5).mean(), color='red')
plt.plot(sdf['frame_index'], sdf['total_objects'].rolling(5).mean(), color='blue')
```
